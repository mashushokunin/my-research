```python
from pathlib import Path
import json
import pandas as pd

def find_project_root(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / 'data').exists() and (candidate / 'scripts').exists():
            return candidate
    raise RuntimeError('Run this notebook from inside the my-research project.')

project_root = find_project_root(Path.cwd().resolve())
frames_root = project_root / 'data' / 'interim' / 'frames_10fps'
rows = []
for metadata_path in sorted(frames_root.rglob('metadata.json')):
    metadata = json.loads(metadata_path.read_text())
    sequence = metadata_path.parent.relative_to(frames_root).as_posix()
    rows.append({
        'sequence': sequence,
        'frames': metadata.get('frames_written'),
        'duration_sec': metadata.get('source_duration_sec'),
        'source_fps': metadata.get('source_fps'),
        'target_fps': metadata.get('target_fps'),
        'size': f"{metadata.get('source_width')}x{metadata.get('source_height')}",
    })
pd.DataFrame(rows)
```
